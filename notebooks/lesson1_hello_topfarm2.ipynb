```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])
colors = ['blue', 'red', 'green', 'yellow']
labels = ['T1', 'T2', 'T3', 'T4']

plt.figure(figsize=(6, 4))
for i in range(len(x)):
    plt.scatter(x[i], y[i], color=colors[i], label=labels[i])
    plt.annotate(labels[i], (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title("Layout turbin – Zajęcia 1 z kolorami i etykietami")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
